# Proceso de EDA:
1- Cargamos los datos

```python

#penetracion_hogares: penetración/ingreso de internet fijo a los hogares (por provincia, por cada 100 hogares) en Argentina
penetracion_hogares = pd.read_csv('/content/drive/MyDrive/Proyecto2_Data_Analitycs/csv/Internet_Penetracion.csv', sep=',', encoding='utf-8')
```

```python
#acceso_bancha_bangosta: Acceso trimestral a banda ancha y banda angosta por provincia en Argentina 
acceso_bancha_bangosta= pd.read_csv('/content/drive/MyDrive/Proyecto2_Data_Analitycs/csv/Internet_BAF.csv', sep=',', encoding='utf-8')
```

```python
# acceso_x_tecnologia_prov: Acceso a internet fijo por tecnologia y provincia de Argentina
acceso_x_tecnologia_prov=pd.read_csv('/content/drive/MyDrive/Proyecto2_Data_Analitycs/csv/Internet_Accesos-por-tecnologia.csv', sep=',', encoding='utf-8')
```

```python
#acceso_total_x_tecnologia: Acceso trimestral a internet fijo por tecnologia a nivel nacional
acceso_total_tecnologia=pd.read_csv('/content/drive/MyDrive/Proyecto2_Data_Analitycs/csv/Internet_Accesos-por-tecnologia (1).csv', sep=',', encoding='utf-8')
```

2- Procedemos a limpiar los datos 


```python
# Verificamos que no existan valores nulos o faltantes.
penetracion_hogares.isna().sum()
penetracion_hogares
```

```python
#Chequeamos el tipo de dato de cada columna
penetracion_hogares.info()
```

```python
acceso_bancha_bangosta.isna().sum()
```

```python
#Encontramos dos columnas vacias en 'Dialup', por lo tanto procedemos a completar la fila con '0'
acceso_bancha_bangosta['Dial up'].fillna(0, inplace=True)
acceso_bancha_bangosta
```

```python
#Continuamos con el siguiente datasets para verificar si hay valores nulos 
acceso_x_tecnologia_prov.isna().sum()
acceso_x_tecnologia_prov
```

```python
#Notamos que la fila 840 contiene datos que no son relevantes, por lo tanto procedemos a eliminarla
acceso_x_tecnologia_prov = acceso_x_tecnologia_prov.drop(840)
```

```python
#Chequeamos los tipo de datos del dataset acceso_bancha_bangosta
acceso_bancha_bangosta.info()
```

```python
#Notamos que hay valores str que deberian ser int64, por lo tanto pasamos los valores a enteros
acceso_bancha_bangosta['Banda ancha fija'] = acceso_bancha_bangosta['Banda ancha fija'].str.replace(r'\D', '')
acceso_bancha_bangosta['Banda ancha fija'] = acceso_bancha_bangosta['Banda ancha fija'].astype('int64')
acceso_bancha_bangosta['Total'] = acceso_bancha_bangosta['Total'].str.replace(r'\D', '')
acceso_bancha_bangosta['Total'] = acceso_bancha_bangosta['Total'].astype('int64')
```

```python
#Notamos que hay valores str que deberian ser int64, por lo tanto pasamos los valores a enteros
acceso_x_tecnologia_prov['Año'] = acceso_x_tecnologia_prov['Año'].str.replace(r'\D', '')
acceso_x_tecnologia_prov['Año'] = acceso_x_tecnologia_prov['Año'].astype('int64')
acceso_x_tecnologia_prov['Trimestre'] = acceso_x_tecnologia_prov['Trimestre'].str.replace(r'\D', '')
acceso_x_tecnologia_prov['Trimestre'] = acceso_x_tecnologia_prov['Trimestre'].astype('int64')
acceso_x_tecnologia_prov['ADSL'] = acceso_x_tecnologia_prov['ADSL'].str.replace(r'\D', '')
acceso_x_tecnologia_prov['ADSL'] = acceso_x_tecnologia_prov['ADSL'].astype('int64')
acceso_x_tecnologia_prov['Cablemodem'] = acceso_x_tecnologia_prov['Cablemodem'].str.replace(r'\D', '')
acceso_x_tecnologia_prov['Cablemodem'] = acceso_x_tecnologia_prov['Cablemodem'].astype('int64')
acceso_x_tecnologia_prov['Fibra óptica'] = acceso_x_tecnologia_prov['Fibra óptica'].str.replace(r'\D', '')
acceso_x_tecnologia_prov['Fibra óptica'] = acceso_x_tecnologia_prov['Fibra óptica'].astype('int64')
acceso_x_tecnologia_prov['Total'] = acceso_x_tecnologia_prov['Total'].str.replace(r'\D', '')
acceso_x_tecnologia_prov['Total'] = acceso_x_tecnologia_prov['Total'].astype('int64')

```

```python
# Unimos Los datasets 'acceso_x_tecnologia_prov' y 'acceso_bancha_bangosta' con la funcion merge, para luego poder trabajar en power bi con una sola tabla
acceso_internet_x_tecnologia = pd.merge(acceso_x_tecnologia_prov, acceso_bancha_bangosta, on= ['Provincia', 'Trimestre', 'Año'], how= 'left')
acceso_internet_x_tecnologia
```
